## Decision Trees

Decision Trees are a non-parametric supervised learning method used for classification (DecisionTreeClassifier) and regression(DecisionTreeRegressor).

The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features.

The algorithm is popular when the decision boundary is non-linear and complex. It is also the algorithm that provides clear decision criteria under which classification is made.

Suppose we have the following dataset with two artificial classes. The tree aims to learn series of test questions in the form of if/else statements and eventually come up with the correct classification.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.datasets import make_moons
from sklearn.tree import DecisionTreeClassifier, plot_tree

%matplotlib inline
plt.style.use('seaborn-whitegrid')


X, y = make_moons(n_samples=100, noise=0.2)

df = pd.DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue'}
fig, ax = plt.subplots(figsize=(6, 4))
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
plt.show()
```

We fit the Decision Tree Classifier with maximum depth of one for the tree.

The decision tree is the structured plan to get a set of attributes to *test* in order to predict an output. It consists of three types of nodes:

- decision nodes (two or more branches)
- leaf nodes (algorithm decisions)
- root node

```python
clf = DecisionTreeClassifier(max_depth=1)
clf = clf.fit(X, y)
plt.figure(figsize=(4, 6))
plot_tree(clf.fit(X, y)) 
```

In each decision node we have an if/else statement for some attribute. If the result of this statement is Yes we go to the left node, if No then to the right node.

The root node represents the whole dataset. It has 100 samples which are equally split into two classes.

Gini refers to the evaluation metric for the algorithm performance. It measure the probability of a random sample being classified incorrectly if we would randomly pick a label according to the distribution among classes in the branch.

Samples indicates the number of examples in each node and value is a vector with the number of examples within each class.

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)



plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, s=20,edgecolor='k')
plt.xlabel('X[0]'); plt.ylabel('X[1]'); plt.title('Decision boundary of tree with depth one')
```

So the algorithm picks up the *test* whether X[1] <= 0.38 as it thinks that it best splits the two classes and yields the most information for the target variable.

If the test is true the point is assigned to the left node, which contains 11 points belonging to the class 0 and 47 points belonging to the class 1. The decision boundary is not perfect since there are points in class 0 below it and class 1 above it.

We increase tree's depth, fit the model and draw a decision boundary.

```python
clf = DecisionTreeClassifier(max_depth=2)
clf = clf.fit(X, y)
plt.figure(figsize=(6, 8))
plot_tree(clf.fit(X, y)) 
```

The process repeats and we get the binary tree of decisions. Since we are dealing with the one feature at a time we receive horizontal and vertical splits along the axes.

```python
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, s=20,edgecolor='k')
plt.xlabel('X[0]'); plt.ylabel('X[1]'); plt.title('Decision boundary of tree with depth two')
```

Now we are not restricting the maximum depth of the tree. The algorithms runs until it will end up with the leaves that contains target value that belong to the single class. Plot the decision tree and the decision boundary.

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X, y)
plt.figure(figsize=(10, 14))
plot_tree(clf.fit(X, y)) 
```

```python
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, s=20,edgecolor='k')
plt.xlabel('X[0]'); plt.ylabel('X[1]'); plt.title('Decision boundary of tree with depth 7')
```

Predicitons for the new unseen data points is based on the partition of the feature space where the point falls. Then prediction is either the majority target class (if multiple classes occur) or the single target class.

For the regression tasks decision tree works in the same manner. We follow along the tree nodes based on the tests until we reach the leaf in which point falls. The predicted value is the mean target of the training points in the leaf. 

Decision Trees suffer from model overfitting as they are 100% sure as they reach the leaves with points belong to the single class.

There are two strategies to decrease model overfitting.

1) Pre-pruning, stopping the creation of the tree earlier by setting the maximum depth of the tree, limiting the maximum number of leaves, or holding the minimum number of points in a node to keep further split.

2) Pruning, collapsing the nodes with little of information.


Let's Decision Tree classifier performance on the breast cancer dataset.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
cancer.data, cancer.target, stratify=cancer.target, random_state=42)
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

The training set accuracy is 100%, because the tree is large and memorized all the labels on the training data. The test set accuracy is a bit less than for the linear models we used to fit the same dataset.

To decrease model overfitting we set the max depth for the tree equal to 4. It will allow to provide 4 tests and then stops tree from growing.

```python
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)
print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

We achieved better model performance on the test set by pruning the tree in advance.

We can use Graphviz utility to draw the decision tree in more readable way. We assign feature names, prediction of the class for each leaf. The nodes are colored with respect to the certainty in prediction of two classes - malignant and benign tumor.

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file="tree.dot", class_names=["malignant", "benign"], feature_names=cancer.feature_names, 
                impurity=False, filled=True)

import graphviz
with open("tree.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

#### Decision Tree benefits and limitations

- the algorithms provides intuitive and easy to read output which can be used to easily convey clear rules that were used to make predictions;
- there is no need to scale features before algorithm implementation;
- it applies to both Regression and Classification problems;
- it works well with quantitative and qualitative predictors;

BUT

It is subject to severe model overfitting (partially the problem can be solved with pruning of the tree OR usage of Ensemble model like Random Forest or Gradient Boosting Decision Tree).
